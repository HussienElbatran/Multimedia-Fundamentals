# Multimedia Fundamentals
  
 This notebook gives a simplified explanation of core multimedia concepts used in software systems.

## 1. What is Multimedia?
 
 Multimedia means combining multiple types of digital content in one system:
 - Text
 - Images
 - Audio
 - Video
 
 Examples: YouTube, Instagram reels, online games, e-learning apps, video conferencing apps.

## 2. Images (Still Media)
 
 A digital image is a grid of pixels. Each pixel stores color information.
 
 Important ideas:
 - **Resolution**: number of pixels (e.g., 1920x1080).
 - **Color depth**: bits per pixel (e.g., 24-bit RGB).
 - **Formats**: JPEG (lossy), PNG (lossless), GIF (limited colors).
 
 Use lossy formats when small file size is important; use lossless formats when quality must be preserved.

## 3. Audio
 
 Digital audio is created by sampling an analog sound wave.
 
 Important ideas:
 - **Sampling rate**: samples per second (e.g., 44.1 kHz).
 - **Bit depth**: precision of each sample (e.g., 16-bit).
 - **Channels**: mono (1), stereo (2).
 - **Formats**: WAV (usually uncompressed), MP3/AAC/OGG (compressed).

## 4. Video
 
 Video is a sequence of images (frames) shown quickly, usually with audio.
 
 Important ideas:
 - **Frame rate**: frames per second (fps), e.g., 24, 30, 60.
 - **Resolution**: e.g., 1280x720, 1920x1080.
 - **Codec**: method to compress/decompress media (H.264, H.265, VP9, AV1).
 - **Container**: file wrapper (MP4, MKV, WebM) holding video/audio/subtitles.

## 5. Compression (Why it matters)
 
 Multimedia files are large. Compression reduces storage and transfer cost.
 
 - **Lossless compression**: no quality loss after decompression (e.g., PNG, FLAC).
 - **Lossy compression**: removes less important details to save more space (e.g., JPEG, MP3, H.264).
 
 Trade-off: smaller file size vs. quality.

## 6. Real-World Pipeline
 
 Typical multimedia app pipeline:
 1. Capture (camera/microphone)
 2. Encode/compress
 3. Store or stream
 4. Decode on client
 5. Render/playback
 
 This pipeline is used in video calls, streaming platforms, and smart surveillance systems.
